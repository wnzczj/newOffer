## 链表结构重组

```python
"""
题型1：单链表反转
解题思路：解题方法有3种
（1）借助栈，将链表依次放入栈中，然后借助栈的先进后出的特点，重新建立链表
     缺点：空间复杂度为 n ，浪费空间
（2）使用递归方法：（本质也是借用栈）
     单链表节点数：n，状态：dp(n)
     转移方程：反转(节点n,dp(n-1)) 其中dp(n-1)表示，将除第一个节点外其他节点看成一个整体===》等价于反转两个节点
     缺点：时间复杂度高
（3）使用3个指针进行反转
    1）先从 2 个节点的反转讲：有单链表 a-->b ，则反转只需要 2 个指针 start 和 end，则start=a，end=b
    则，start = end.next 完成反转
    2）如果单链表含有 3 个以上节点，则会出现错误。有单链表 a-->b-->c ,
    如果依然安装 1）中方法，则 start = a，end = b，然后 start = end.next 结果：a-->b-->c 变为 a<--b  c
    错误：当继续执行：start = b，end = ? 会发现我们找不到 b 的下一个节点
    3）针对 2）的解决办法，使用 3 个指针，start，mid，end，其中start，mid还是执行 1）中操作，增加 end 指向第 3 个节点
    则：单链表：a-->b-->c-->d，则start=a，mid=b，end=c，则进行 1）操作，则 start=end.next ,变为 a<--b  c-->d
    然后，关键点：end保存着节点 b 的下一个节点，则：start=b，mid=end，end=end.next，对start和mid继续进行 1）操作，然后进行循环
（4）补充方法：最初接触数据结构时，使用list保存所有节点，然后reserved(list)进行重新构建节点，本质上跟方法（1）借助栈是一样的
"""
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, x):
        self.val = x
        self.next = None

class Solution(object):
    def __init__(self):
        self.finalNode = None
    
    # 方法1：借助栈
    def reverseList_1(self, head):
        if not head:
            return None
        # 建立栈
        stack = []
        while head:
            stack.append(head)
            head=head.next
        """
        注意：要先将所有的节点的指针指向None
        因为如果不这样会导致：1-->2-->3-->4-->5 变为 5-->4-->3-->2-->1-->2-->1-->2-->1-->2-->1...
        导致运行时间很长
        放在LeetCode和牛客不能accept
        """
        for i in stack:
            i.next = None
        # 出栈重新建立链表
        first = stack.pop()
        temp = first
        while len(stack)>0:
            temp.next=stack.pop()
            temp = temp.next
        return first
    
    # 方法2：递归方法解决
    def reverseList(self, head):
        if head is None:
            return None
        self.digui(head)
        return self.finalNode
    
    # 使用递归来出来单链表
    def digui(self,head):
        # 结束条件--最后一个节点停下来
        if head.next is None:
            # 将此时的节点保存下来
            self.finalNode = head
            return head
        # 转移条件
        # 交换两部分
        start = head
        end = self.digui(head.next)
        start.next = None
        end.next = start
        """
        注意：此处返回的一定是 head
        因为：此时的head是已经处理好的"子问题"的结果
        例如：
        """
        return head
    
    # 方法3：使用 3 个指针来做
    def reverseList(self, head):
        if head is None:
            return 
        if head.next is None:
            return head
        if head.next.next is None:
            start = head
            end = head.next
            start.next = None
            end.next = start
            return end
        # 使用 3 个指针保存 3 个节点
        start = head
        mid = head.next
        end = head.next.next
        start.next = None
        # 依次进行交换
        while end.next is not None:
            # 重新调整这 3 个
            mid.next = start
            # 移动这 3 个指针
            start = mid
            mid = end
            end = end.next
        mid.next = start
        end.next = mid
        return end
```

```python
# 创建方法
class Solution:
    # 链表反转
    # 注意情况：[]
    def ReverseList(self, pHead):
        # write code here
        if pHead is None:
            return None
        first = []
        # 收集节点
        while pHead is not None:
            first.append(pHead)
            pHead = pHead.next
        for i in first:
            i.next = None
        # 反转
        for i in range(len(first)-1,0,-1):
            first[i].next = first[i-1]
        return first[-1]
    
    # 删除排序链表中重复的节点
    # 注意情况：
    # 空：[]
    # 全是重复节点：[1,1,1,1,1]
    def deleteDuplication(self, pHead):
        if pHead is None:
            return None
        # 收集链表节点,获得对应值和次数
        first = []
        while pHead is not None:
            # 分别对应：[节点，节点值，次数]
            first.append([pHead,pHead.val,0])
            pHead = pHead.next
        # 初始化节点值次数词典
        second = {}
        for i in first:
            second[i[1]] = 0
        # 获得节点出现次数
        for i in first:
            if i[1] in second.keys():
                second[i[1]] += 1
        # 选取重复节点
        third = []
        for i in second.keys():
            if second[i] > 1:
                third.append(i)
        # 获得节点
        forth = []
        for i in first:
            if i[1] not in third:
                forth.append(i[0])
        # 如果全是重复节点
        if len(forth) == 0:
            return None
        # 重建链表
        for i in forth:
            i.next = None
        for i in range(len(forth)-1):
            forth[i].next = forth[i+1]
        return forth[0]
    
```
