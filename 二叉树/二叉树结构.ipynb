```python
# 定义二叉树
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
# 定义栈
class Stack:
    def __init__(self):
        self.val = []
    def push(self,x):
        self.val.append(x)
    def pop(self):
        self.val.pop()
# 定义队列
class Queue:
    def __init__(self):
        self.val = []
    def push(self,x):
        self.val.append(x)
    def pop(self):
        popVal = self.val.pop(0)
        return popVal
```

```python
# -*- coding:utf-8 -*-
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None

class Solution:
    def __init__(self):
        self.temp = 1
    # 获取二叉树子结构
    # 思路：
    # （1）查看根节点是否一样，如果一样，
    # （2）则：继续判断左子树和右子树
    # （3）如果左子树或右子树不同，则
    # （4）继续从子树的根节点（pNode2）开始匹配
    # （5）如果一开始就根节点不同，则查看左子树，右子树
    def HasSubtree(self,pNode1,pNode2):
        if pNode2 is None and self.temp >1:
            return True
        if pNode2 is None:
            return False
        if pNode1 is None or pNode1.val != pNode2.val:
            return False
        leftResult1,leftResult2,rightResult1,rightResult2,leftResult11,rightResult11 = False,False,False,False,False,False
        # 查看根节点
        if pNode1.val == pNode2.val:
            self.temp += 1
            leftResult1 = self.HasSubtree(pNode1.left,pNode2.left)
            if leftResult1 == False:
                leftResult11 = self.HasSubtree(pNode1.left,pNode2)
            rightResult1 = self.HasSubtree(pNode1.right,pNode2.right)
            if rightResult1 == False:
                rightResult11 = self.HasSubtree(pNode1.right,pNode2)
        else:
            leftResult2 = self.HasSubtree(pNode1.left,pNode2)
            rightResult2 = self.HasSubtree(pNode1.right,pNode2)
        return (leftResult1 and rightResult1) or leftResult2 or rightResult2 or leftResult11 or rightResult11
    
    # 二叉树的镜像
    # 思路：根节点不动，左右子树交换
    def mirror(self, pNode):
        if pNode is None:
            return None
        temp = pNode.left
        pNode.left = pNode.right
        pNode.right = temp
        self.mirror(pNode.left)
        self.mirror(pNode.right)
        return pNode
    
    # 二叉树复制
    # 思路：
    # 第（1）步：先复制根节点
    # 第（2）步：复制左子树，再复制右子树，并把左子树，右子树分别添加为复制的根节点的左右子树
    def cloneTree(self, root):
        # write your code here
        if root is None:
            return None
        newroot = TreeNode(root.val)
        if root.left:
            newroot.left = self.cloneTree(root.left)
        if root.right:
            newroot.right = self.cloneTree(root.right)
        return newroot
    
    # 对称二叉树
    # 定义：二叉树的镜像与原二叉树相同，则为对称二叉树
    # 思路：
    # 第（1）步：将原来二叉树复制一份
    # 注意！！：此处不能里面函数传进来的pNode跟与它镜像后等到的mirrorPNode，比较相同子结构
    # 因为：在pNode进行镜像时，它的结构已经改变，也即：mirrorPNode = 新的pNode，即pNode镜像改变后赋值给mirrorPNode
    # 第（2）步：求二叉树的镜像
    # 第（3）步：根据前面二叉树判断子结构，判断二叉树镜像与原二叉树是否相同，如果相同则为对称二叉树，否则不是
    def isSymmetrical(self,pNode):
        if pNode is None:
            return True
        # 保存原来二叉树
        oriNode = self.cloneTree(pNode)
        # 获得二叉树镜像
        mirrorPNode = self.mirror(pNode)
        # 判断子结构
        isResult = self.HasSubtree(oriNode,mirrorPNode)
        return isResult 
    
    # 二叉树的序列化与反序列化
    # 思路：此处由于算上空节点，则进行反序列化时，只需要把二叉树看成“满二叉树”来进行处理。（很重要！！！）
    # 如果遇到空节点“#”，不进入 if 判断，则root = None返回
    # 序列化--前序遍历
    def frontNew(self, pNode):
        if pNode is None:
            return '#'
        # self.frontLis.append(pNode.val)
        left = self.frontNew(pNode.left)
        right = self.frontNew(pNode.right)
        # 节点间以“！”相隔，空节点用“#”代替
        return list(str(pNode.val)) + ['!'] + list(left) + ['!'] + list(right)
    # 序列化
    def Serialize(self, root):
        # 前序遍历
        first = self.frontNew(root)
        # 序列化
        first = ''.join(first)
        # 反序列化
        return first
    # 反序列化
    def Deserialize(self, s):
        first = s.split("!")
        second = self.deser(first)
        return second
    # 反序列化具体方法
    def deser(self,data):
        if len(data) <=0:
            return None
        # 获取根节点
        first = data.pop(0)
        root = None
        if first != '#':
            # 注意！！！：
            # 此处：错误代码：root = TreeNode(first)
            # 因为此处，节点值需要是int型
            # 为什么？？
            root = TreeNode(int(first))
            root.left = self.deser(data)
            root.right = self.deser(data)
        return root
```
