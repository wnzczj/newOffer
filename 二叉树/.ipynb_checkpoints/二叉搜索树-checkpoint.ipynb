```python
# 定义二叉树
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
# 定义栈
class Stack:
    def __init__(self):
        self.val = []
    def push(self,x):
        self.val.append(x)
    def pop(self):
        self.val.pop()
# 定义队列
class Queue:
    def __init__(self):
        self.val = []
    def push(self,x):
        self.val.append(x)
    def pop(self):
        popVal = self.val.pop(0)
        return popVal
```

```python
class Solution():
    def __init__(self):
        self.midLis = []
        self.temp = 1
    # 中序遍历
    def mid(self, pNode):
        if pNode is None:
            return
        self.mid(pNode.left)
        self.midLis.append(pNode)
        self.mid(pNode.right)

    # 将二叉搜索树--》排序双向链表
    def Convert(self, pNode):
        if pNode is None:
            return 
        # 中序遍历：
        self.mid(pNode)
        # 将中序遍历结果转化为双向链表
        # 从左往右添加right指针
        for i in range(len(self.midLis) - 1):
            self.midLis[i].right = self.midLis[i + 1]
        # 从右往左添加left指针
        for j in range(len(self.midLis) - 1, 0, -1):
            self.midLis[j].left = self.midLis[j - 1]
        return self.midLis[0]
    
    # 二叉搜索树第k个结点
    def KthNode(self, pNode,k):
        if pNode is None:
            return
        # 中序遍历：
        self.mid(pNode)
        if k <= 0 or k > len(self.midLis):
            return
        return self.midLis[k-1]
    
    # 二叉搜索树的后序遍历序列
    # 情况分为：
    # 正常情况： [4,8,6,12,16,14,10]
    # 只有左子树：[1,2,3,4,5]
    # 只有右子树：[5,4,3,2,1]
    # 一个元素：[5]
    # 空元素：[]
    # 注意：空元素[]，要注意跟使用递归时，递归结束条件相区别
    # 方法，增加运行次数标记：self.temp
    # 注意：self.temp 不要放在 for 循环或 if 判断语句内，因为可能 for 的range(len(sequence)) 中 len(sequence) == 0
    # 导致：for 循环不被运行，或 if 语句不满足判断条件，也不会运行
    # 见下面：for 循环处
    def VerifySquenceOfBST(self, sequence):
        if sequence == [] and self.temp > 1:
            return True
        if sequence == []:
            return False
        # 思路：
        # 根据最后一个元素划分数组为两部分
        # 左边 < 该元素; 右边 > 该元素
        first = sequence.pop()
        second = 0
        third = True
        if len(sequence) == 1 and sequence[0] > first:
            second = -1
        if len(sequence) == 1 and sequence[0] < first:
            second = 0
        for i in range(len(sequence)-1):
            # 注意：原来 self.temp += 1 放在此处，但没有考虑到 for 循环可能不会运行，导致代码出错
            if sequence[i] < first and sequence[i+1] > first:
                second = i
                break
            if i == len(sequence)-2 and sequence[i + 1] < first:
                second = i + 1
                break
            if i == 0 and sequence[i] > first:
                second = i - 1
                break
        # 最终：self.temp += 1放在此处，代码正确
        self.temp += 1
        left = sequence[:second + 1]
        right = sequence[second + 1:]
        for i in left:
            if i > first:
                third = False
        for i in right:
            if i < first:
                third = False
        leftResult = self.VerifySquenceOfBST(left)
        rightResult = self.VerifySquenceOfBST(right)
        return third and leftResult and rightResult
    
    
```

```python
# 搭建树结构
n10 = TreeNode(10)
n6  = TreeNode(6)
n14 = TreeNode(14)
n4  = TreeNode(4)
n8  = TreeNode(8)
n12 = TreeNode(12)
n16 = TreeNode(16)

n10.left  = n6
n10.right = n14
n6.left   = n4
n6.right  = n8
n14.left  = n12
n14.right = n16
```
