```python
# 定义二叉树
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
# 定义栈
class Stack:
    def __init__(self):
        self.val = []
    def push(self,x):
        self.val.append(x)
    def pop(self):
        self.val.pop()
# 定义队列
class Queue:
    def __init__(self):
        self.val = []
    def push(self,x):
        self.val.append(x)
    def pop(self):
        popVal = self.val.pop(0)
        return popVal
```

```python
# 前序遍历
# 定义解决方法
class Solution:
    def __init__(self):
        # 前序结果
        self.frontLis = []
        # 中序结果
        self.midLis = []
        # 后序结果
        self.lasLis = []
        # 层次遍历
        self.cengLis = []
        # 之字形
        self.zhiLis = []
    # 前序遍历
    def front(self, pNode):
        if pNode is None:
            return
        self.frontLis.append(pNode.val)
        self.front(pNode.left)
        self.front(pNode.right)
    # 中序遍历
    def mid(self, pNode):
        if pNode is None:
            return
        self.mid(pNode.left)
        self.midLis.append(pNode.val)
        self.mid(pNode.right)
    # 后序遍历
    def las(self, pNode):
        if pNode is None:
            return
        self.las(pNode.left)
        self.las(pNode.right)
        self.lasLis.append(pNode.val)
    # 按层遍历
    # 使用队列进行遍历
    def ceng(self,pNode):
        # 建立队列
        queue = Queue()
        # 根节点入队
        queue.push(pNode)
        while(len(queue.val)>0):
            # 出队
            first = queue.pop()
            self.cengLis.append(first.val)
            # 入队
            if first.left is not None:
                queue.push(first.left)
            if first.right is not None:
                queue.push(first.right)
    # 之字顺序遍历
    def zhi(self,pNode):
        if pNode is None:
            return []
        # 建立队列
        queue = Queue()
        # 根节点入队
        queue.push([pNode,0])
        self.zhiLis.append([pNode.val,1])
        while(len(queue.val)>0):
            # 出队
            first = queue.pop()
            if first[0] != pNode:
                self.zhiLis.append([first[0].val,first[1]+1])
            # 入队
            if first[0].left is not None:
                queue.push([first[0].left,first[1]+1])
            if first[0].right is not None:
                queue.push([first[0].right,first[1]+1])
        # 将每层的节点进行整合
        second = []
        for i in range(1,self.zhiLis[-1][1]+1):
            third = []
            for j in self.zhiLis:
                if j[1] == i:
                    third.append(j[0])
            if i % 2 == 0:
                third = list(reversed(third))
            second.append(third)
        self.zhiLis = second
        
    # 从上往下，按层打印二叉树，每层，从左往右
    # 注：也即跟上面“之字形”打印中，去掉下面即可
    # if i % 2 == 0:
    #     third = list(reversed(third))
    
    # 注意：
    # 此处树为：{8,6,10,5,7,9,11}
    # 中序遍历结果为：[5,6,7,8,9,10,11]
    # 特殊情况：当pNode = 11，则根据self.midLis来获取时，要注意越界，见下
    # 解析：因为此处只给了目标节点：pNode，则需要根据父指针，找到根节点
    # 然后再确定该节点 pNode 的下一个节点
    def GetNext(self, pNode):
        first = pNode
        # 找到根节点
        while first.next is not None:
            first = first.next
        # 中序遍历
        self.mid(first)
        # 注意：pNode为11的情况
        if self.midLis.index(pNode)+1 >= len(self.midLis):
            return
        second = self.midLis[self.midLis.index(pNode)+1]
        # # 获取下一个节点
        return second
    
    
```
