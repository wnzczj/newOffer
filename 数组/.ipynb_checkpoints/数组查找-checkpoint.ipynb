```python
class Solution():
    # 数组中只出现1次的数字
    def FindNumsAppearOnce(self, array):
        if array is []:
            return []
        # write code here
        # 获得数字字典
        first = list(set(array))
        # 给每个数字添加次数
        second = {}
        for i in first:
            second[i] = 0
        # 获取次数
        for i in array:
            second[i] += 1
        # 获取只出现一次的数字
        third = []
        for i in second.keys():
            if second[i] == 1:
                third.append(i)
        return third
```

```python
# 测试
data = [1,1,2,3,4,2,3,5,4,6,5,7]
s = Solution()
result = s.FindNumsAppearOnce(data)
print(result)
```
